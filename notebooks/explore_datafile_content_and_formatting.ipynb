## Notes on the content - old vs. new CSV files
Newer CSV files contain 5 new columns. Mostly for currency conversions. The old-style conversion using static_usd_rate is close to the newer fx_rate and works OK. Current currency is always 'USD' and usd_type is 'domestic'. The new columns don't seem necessary. is_starrable is mostly False, not sure how it could be used.
